```python
import librosa
import librosa.display

import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt
```

```python
bell = np.load('../augmented_data/mixed_bells.npy')
noise = np.load('../augmented_data/noise.npy')

sr = 22050
```

```python
S = librosa.feature.melspectrogram(noise[1], sr=sr, n_mels=128)
S_DB = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S_DB, sr=sr, x_axis='time', y_axis='mel');
plt.colorbar(format='%+2.0f dB');
```

```python
S = librosa.feature.melspectrogram(bell[1], sr=sr, n_mels=128)
S_DB = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S_DB, sr=sr, x_axis='time', y_axis='mel');
plt.colorbar(format='%+2.0f dB');
```
